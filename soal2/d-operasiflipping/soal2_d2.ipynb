```python
from PIL import Image
import matplotlib.pyplot as plt

src = "../../static/img/butterfly.jpeg"
img = Image.open(src)

print("Original Image")
plt.imshow(img)
plt.show()

print("Flipping Vertikal")
plt.imshow(img.transpose(Image.FLIP_TOP_BOTTOM))
plt.show()

print("Flipping Horizontal")
plt.imshow(img.transpose(Image.FLIP_LEFT_RIGHT))
plt.show()

print("Flipping Titik Asal")
plt.imshow(img.transpose(Image.ROTATE_180))
plt.show()
```
