```python
import matplotlib.pyplot as plt
import cv2

src = "../../static/img/butterfly.jpeg"
img = cv2.imread(src)
img_gray = cv2.imread(src, 0)

t = 128
H, W = img_gray.shape[:2]

img_brighten = img_gray.copy()
bright = 70

for i in range(H):
    for j in range(W):
        pixel = img_brighten[i, j] + bright
        if pixel > 255:
            img_brighten[i, j] = 255
        else:
            img_brighten[i, j] = pixel

plt.imshow(img)
plt.show()
plt.imshow(img_gray)
plt.show()
plt.imshow(img_brighten)
plt.show()
```
