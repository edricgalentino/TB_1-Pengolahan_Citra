```python
from PIL import Image
import matplotlib.pyplot as plt

src = "../../static/img/butterfly.jpeg"
img = Image.open(src)

print("Original Image")
plt.imshow(img)
plt.show()

print("Rotated 45 degrees")
plt.imshow(img.rotate(45))
plt.show()

print("Rotated -45 degrees")
plt.imshow(img.rotate(-45))
plt.show()

print("Rotated 90 degrees")
plt.imshow(img.rotate(90))
plt.show()

print("Rotated 180 degrees")
plt.imshow(img.rotate(180))
plt.show()

print("Rotated 270 degrees")
plt.imshow(img.rotate(270))
plt.show()

print("Rotated 360 degrees")
plt.imshow(img.rotate(360))
plt.show()
```
